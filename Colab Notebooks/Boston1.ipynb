```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
```

```python
boston = load_boston()
```

```python
data = pd.DataFrame(boston.data)
```

```python
data.columns = boston.feature_names
data.head()
```

```python
data['PRICE'] = boston.target
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
x = data.drop(['PRICE'], axis = 1)
y = data['PRICE']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
model = lm.fit(x_train, y_train)
```

```python
ytrain_pred = lm.predict(x_train)
ytest_pred = lm.predict(x_test) 
```

```python
df = pd.DataFrame(ytrain_pred, y_train)
df = pd.DataFrame(ytest_pred, y_test)
```

```python
plt.scatter(y_train, ytrain_pred, color = "blue")
plt.scatter(y_test, ytest_pred, color = "lightgreen")
plt.xlabel('True values')
plt.ylabel("Predicted")
plt.title("True Value vs Predicted values")
plt.legend(loc = "upper left")
plt.plot()
plt.show()
```
