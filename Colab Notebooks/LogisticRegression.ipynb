```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("Social_Network_Ads.csv")
df
```

```python
dummies = pd.get_dummies(df["Gender"])
merged = pd.concat([df, dummies], axis='columns')
df = merged.drop(["Female", "Gender"], axis='columns')
df.head()
```

```python
df.isnull().sum()
```

```python
x=df.iloc[:,:-1]
y=df['Purchased']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
lg.fit(x_train,y_train)
```

```python
prediction_x_test = lg.predict(x_test)
prediction_x_test
```

```python
prediction_x_train = lg.predict(x_train)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
print(accuracy_score(y_test,prediction_x_test))
```

```python
print(confusion_matrix(y_test,prediction_x_test))
```

```python
print(classification_report(y_test,prediction_x_test))
```
