```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Iris.csv")
df.head()
```

```python
fig = plt.figure(figsize = (10,5))
ax = fig.gca()
plt.title("BoxPlot of All Features")
df.boxplot(ax = ax)
```

```python

fig = plt.figure(figsize = (10,10))
ax = fig.gca()
plt.title("Histogram of All Features")
df.hist(ax = ax)
```
